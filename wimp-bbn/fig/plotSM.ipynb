```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_table('plotSM.dat', sep=' ')
df.drop('Unnamed: 8', axis = 1, inplace = True)
```

```python
real = df[df['No'] == 1].copy()
comp = df[df['No'] == 2].copy()
Majo = df[df['No'] == 3].copy()
Dirc = df[df['No'] == 4].copy()
```

```python
plt.figure(1)
plt.xscale('log')
plt.yscale('linear')
plt.xlabel(r'$m_\chi$')
plt.ylabel(r'$Y_p$')
plt.xlim(0.01,100)
plt.ylim(0.235, 0.3)
plt.plot(real['mass'],real['Yp'],'b--', label='real')
plt.plot(comp['mass'],comp['Yp'],'g--', label='real')
plt.plot(Majo['mass'],Majo['Yp'],'k-', label='real')
plt.plot(Dirc['mass'],Dirc['Yp'],'r-', label='real')
plt.savefig('SM1.pdf')
```

```python
plt.figure(2)
plt.xscale('log')
plt.yscale('linear')
plt.xlabel(r'$m_\chi$')
plt.ylabel(r'$H2/H$')
plt.xlim(0.01,100)
plt.plot(real['mass'],real['H2/H'],'b--', label='real')
plt.plot(comp['mass'],comp['H2/H'],'g--', label='real')
plt.plot(Majo['mass'],Majo['H2/H'],'k-', label='real')
plt.plot(Dirc['mass'],Dirc['H2/H'],'r-', label='real')
plt.savefig('SM2.pdf')
```
